```python
# Passo 1: Importando nossas ferramentas
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Linha especial para garantir que as imagens apaream dentro do notebook
%matplotlib inline

# --- CONFIGURAO ---
# Coloque o nome exato da sua imagem de formulrio de teste aqui
NOME_DO_ARQUIVO_DE_IMAGEM = 'teste.png'

# --- CARREGANDO A IMAGEM ---
# Carrega a imagem usando o OpenCV
img_original = cv2.imread(NOME_DO_ARQUIVO_DE_IMAGEM)

# Verifica se a imagem foi carregada corretamente
if img_original is None:
    print(f"ERRO: No foi possvel carregar a imagem em '{NOME_DO_ARQUIVO_DE_IMAGEM}'")
    print("Verifique se o nome do arquivo est correto e se ele est na mesma pasta que este notebook.")
else:
    print("Imagem original carregada com sucesso!")
    # Converte a cor de BGR (padro do OpenCV) para RGB para exibio correta
    img_rgb = cv2.cvtColor(img_original, cv2.COLOR_BGR2RGB)
    
    # Exibe a imagem original
    plt.figure(figsize=(10, 10)) # Aumenta o tamanho da figura para melhor visualizao
    plt.imshow(img_rgb)
    plt.title("Imagem Original")
    plt.axis('off') # Remove os eixos x e y
    plt.show()
```

```python
# --- EXPERIMENTO 7: Testando uma Eroso Mais Forte ---

# Mantemos o pipeline de Otsu que funcionou para limpar o rudo
img_cinza = cv2.cvtColor(img_original, cv2.COLOR_BGR2GRAY)
img_desfocada = cv2.GaussianBlur(img_cinza, (5, 5), 0)
_, img_otsu = cv2.threshold(
    img_desfocada, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU
)

# PASSO NOVO: Aplicar uma Eroso mais agressiva
# Aumentamos o tamanho do kernel de (2, 2) para (3, 3).
kernel = np.ones((3, 3), np.uint8)
    
# Aplicamos a eroso na imagem j binarizada pelo Otsu
img_final_erodida = cv2.erode(img_otsu, kernel, iterations=1)


print("Resultado com o Mtodo de Otsu + Eroso Forte")
    
# Vamos exibir as duas imagens para comparar o antes e o depois da eroso
plt.figure(figsize=(20, 10))

plt.subplot(1, 2, 1) # Imagem da esquerda
plt.imshow(img_otsu, cmap='gray')
plt.title("Imagem Apenas com Otsu (Letras Grossas)")
plt.axis('off')

plt.subplot(1, 2, 2) # Imagem da direita
plt.imshow(img_final_erodida, cmap='gray')
plt.title("Imagem Final com Eroso Forte (Kernel 3x3)")
plt.axis('off')

plt.show()
```
